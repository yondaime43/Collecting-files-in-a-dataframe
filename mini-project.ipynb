```python
import pandas as pd
import os
```

```python
# looking for necessary path
os.listdir('/home/jupyter-k.sanzhar-13/shared/homeworks/python_ds_miniprojects/4')
```

```python
path = '/home/jupyter-k.sanzhar-13/shared/homeworks/python_ds_miniprojects/4/data'
```

```python
# create df for all data
df = pd.DataFrame()

# read data from all files and add it to general df
for current_path, dirs, files in os.walk(path):
    for file in files:
        # construct full path to file
        data_path = f"{current_path}/{file}"
        
        # read data
        temp_df = pd.read_csv(data_path)
        
        # name, date
        path_parts = current_path.split('/')
        date = path_parts[-2]
        name = path_parts[-1]
        
        # add date and name columns 
        temp_df['date'] = date
        temp_df['name'] = name
        
        # insert data into main df
        df = pd.concat((df, temp_df))
        
        # check for true
        print(df.shape)
```
